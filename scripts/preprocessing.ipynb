```python
import pandas as pd
from sqlalchemy import create_engine

# Database connection
user = 'root'
password = 'tayba'
host = 'localhost'
port = '3306'
db_name = 'football_dwh'
db_connection_str = f'mysql+pymysql://{user}:{password}@{host}:{port}/{db_name}'
engine = create_engine(db_connection_str)

# Read CSV
df = pd.read_csv('datasets/player_season_stats_2018_2019.csv')

# ========== Table 1: players ==========
players = df[['player_name', 'birth_date', 'nationality']].drop_duplicates()
players.insert(0, 'player_id', range(1, len(players) + 1))
players.to_sql('players', con=engine, if_exists='replace', index=False)

# Create player_id mapping
player_map = players.set_index('player_name')['player_id'].to_dict()
df['player_id'] = df['player_name'].map(player_map)

# ========== Table 2: player_seasons ==========
player_seasons = df[[
    'player_id', 'season_id', 'club_name', 'position', 'age',
    'appearances_total', 'minutes_played_total', 'goals_total', 
    'assists_total', 'penalty_goals', 'penalty_misses',
    'clean_sheets_total', 'goals_conceded_total'
]].copy()
player_seasons.rename(columns={'season_id': 'season_year'}, inplace=True)
player_seasons.insert(0, 'season_id', range(1, len(player_seasons) + 1))
player_seasons.to_sql('player_seasons', con=engine, if_exists='replace', index=False)

# ========== Table 3: player_venue_stats ==========
# Home stats
home_stats = df[[
    'player_id', 'minutes_played_home', 'appearances_home',
    'goals_home', 'assists_home', 'clean_sheets_home', 'goals_conceded_home'
]].copy()
home_stats['venue_type'] = 'home'
home_stats.rename(columns={
    'minutes_played_home': 'minutes_played',
    'appearances_home': 'appearances',
    'goals_home': 'goals',
    'assists_home': 'assists',
    'clean_sheets_home': 'clean_sheets',
    'goals_conceded_home': 'goals_conceded'
}, inplace=True)

# Away stats
away_stats = df[[
    'player_id', 'minutes_played_away', 'appearances_away',
    'goals_away', 'assists_away', 'clean_sheets_away', 'goals_conceded_away'
]].copy()
away_stats['venue_type'] = 'away'
away_stats.rename(columns={
    'minutes_played_away': 'minutes_played',
    'appearances_away': 'appearances',
    'goals_away': 'goals',
    'assists_away': 'assists',
    'clean_sheets_away': 'clean_sheets',
    'goals_conceded_away': 'goals_conceded'
}, inplace=True)

# Combine venue stats
venue_stats = pd.concat([home_stats, away_stats], ignore_index=True)
venue_stats.insert(0, 'season_id', list(range(1, len(player_seasons) + 1)) * 2)
venue_stats.insert(0, 'venue_id', range(1, len(venue_stats) + 1))
venue_stats.to_sql('player_venue_stats', con=engine, if_exists='replace', index=False)

# ========== Table 4: player_ranks ==========
player_ranks = df[[
    'player_id', 'rank_league_attackers', 'rank_league_midfielders',
    'rank_league_defenders', 'rank_club_scorer'
]].copy()
player_ranks.insert(0, 'season_id', range(1, len(player_ranks) + 1))
player_ranks.insert(0, 'rank_id', range(1, len(player_ranks) + 1))
player_ranks.to_sql('player_ranks', con=engine, if_exists='replace', index=False)

```

```python
import pandas as pd

def clean_football_data(input_file, output_file):
    """
    Clean and normalize football match data for SQL Pool import
    """
    # Read CSV
    df = pd.read_csv(input_file)

    # Select relevant columns only
    relevant_columns = [
        "Date",
        "HomeTeam",
        "AwayTeam",
        "FTHG",
        "FTAG",
        "FTR",
        "HTHG",
        "HTAG",
        "HTR",
        "HS",
        "AS",
        "HST", "AST"	,
"HF", "AF",	
"HC", 'AC'	,
'HY', "AY"	,
'HR', "AR", 
"B365H", "B365D", "B365A"
    ]
    
    df = df[relevant_columns]

    # Save cleaned data
    df.to_csv(output_file, index=False, date_format='%Y-%m-%d', encoding='utf-8')

    return df
```

```python
import json
import re
import os

# ---- EDIT: keep only these attributes ----
KEEP_KEYS = {
    "Rk", "Player", "Nation", "Pos", "Squad",
    "Age", "Born", "MP", "Starts", "Min", "90s",
    "Gls", "Ast", "CrdY", "CrdR",
    "Gls_1", "Ast_1"
}

# -------------------------------------------

def parse_market_value(raw_val):
    if not raw_val:
        return None, None

    # Example: "â‚¬500k\nLast update: 28/05/2025"
    parts = raw_val.split("\n")
    value = parts[0].replace("â‚¬", "").replace("k", "").replace("m", "").strip()
    value = float(value) 

    # extract last update
    match = re.search(r"(\d{2}/\d{2}/\d{4})", raw_val)
    last_update = match.group(1) if match else None

    return int(value), last_update


def clean_json_file(path_in, path_out):
    with open(path_in, "r", encoding="utf8") as f:
        data = json.load(f)

    cleaned_rows = []

    for row in data:
        new_row = {}

        # Keep only selected fields
        for k in KEEP_KEYS:
            if k in row:
                new_row[k] = row[k]

        # Extract market value if exists
        player_info = row.get("player_info", {})
        market_raw = player_info.get("market_value")

        value, last_update = parse_market_value(market_raw)

        new_row["market_value_â‚¬k"] = value
        new_row["market_value_last_update"] = last_update

        cleaned_rows.append(new_row)

    with open(path_out, "w", encoding="utf8") as f:
        json.dump(cleaned_rows, f, indent=4, ensure_ascii=False)


```

```python
input_folder = "players_info/"
output_folder = "clean_players_info/"

os.makedirs(output_folder, exist_ok=True)

for filename in os.listdir(input_folder):
    if filename.endswith(".json"):
        clean_json_file(
            os.path.join(input_folder, filename),
            os.path.join(output_folder, filename)
        )

print("Cleaning completed!")
```
